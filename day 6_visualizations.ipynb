```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris(as_frame=True)
df=iris.frame

df.head()
```

# Chart 1 - Line Chart
## What this tells us

This chart shows how sepal length changes across the dataset samples.

```python
plt.figure(figsize=(8,5))
plt.plot(df.index, df["sepal length (cm)"],
         color="blue")
plt.title("Sepal Length Across Samples")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.grid(True)
plt.show()
```

# Chart 2 - Bar Chart
## What this tells us

This chart compares the number of samples available for each flower species.

```python
species_count = df["target"].value_counts()
plt.figure(figsize=(6,4))
plt.bar(["Setosa","Versicolor","Virginica"],
        species_count)
plt.title("Number of Flowers per species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

# Chart 3 - Histogram
# What this tells us

This histogram shows how petal lengths are distributed across all flowers.

```python
plt.figure(figsize=(7,5))
plt.hist(df["petal length (cm)"],
         bins=15)
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

# Chart 4 - Boxplot
# What this tells us

This boxplot summarizes petal width and highlights the presence of outliers.

```python
plt.figure(figsize=(6,5))
plt.boxplot(df["petal width (cm)"])
plt.title("Petal Width Boxplot")
plt.ylabel("Width (cm)")
plt.show()
```

# Chart 5 - Scatter Plot
# what this tells us

This scatter plot shows the relationship between sepal length and petal length.

```python
plt.figure(figsize=(7,5))
plt.scatter(df["sepal length (cm)"],
            df["petal length (cm)"])
plt.title("Sepal length vs Petal length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
plt.tight_layout()
```

# Chart 6 - Heatmap
# what this tells us

This heatmap visualizes the correlation strength between the numerical features.

```python
# Without seaborn 
corr = df.drop(columns="target").corr()
plt.figure(figsize=(7,6))
plt.imshow(corr,
           cmap="coolwarm")
plt.colorbar()
plt.xticks(range(len(corr.columns)),
           corr.columns,
           rotation=45)
plt.yticks(range(len(corr.columns)),
           corr.columns)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

### When should you use a bar chart vs. a histogram?
### Bar Chart

Used for categorical data.

Example:

Number of students in each department
Sales by country
Population by city

Example:

Department	Students
CS	120
EE	80
AI	60

Bars are separated.

### Histogram

Used for continuous numerical data.

It shows the distribution of a variable.

Example:

Student ages

18 19 20 20 21 22 22 22 23 24

Histogram groups values into intervals (bins).

Bars touch each other.

### 2. What does a boxplot show that a histogram doesn’t?

A boxplot clearly shows:

Median
Quartiles (Q1 & Q3)
Spread
Outliers

A histogram only shows the frequency distribution.

So if you want to detect outliers, use a boxplot.

### 4. What is a heatmap typically used for in EDA?

A heatmap is used to visualize:

Correlation between variables
Missing values
Large tables of numbers

Example:

Finding which features are highly correlated.

### 5. Why is a pie chart usually a bad choice?

Because humans compare length much better than angles.

With many categories:

difficult to compare

labels overlap

tiny slices become unreadable

Pie charts work best for 3–5 categories only.
